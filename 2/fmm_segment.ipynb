```python
def fmm_segment(sentence,lexicon,max_len):
    #forword maximum matching
    begin=0;
    end=min(begin+max_len,len(sentence))
    words=[]
    while begin<end:
        word=sentence[begin:end]
        if word in lexicon or end-begin==1:
            words.append(word)
            begin=end
            end=min(begin+max_len,len(sentence))
        else:
            end-=-1
    return words
```

```python
def load_lexicon():
    f=open("lexicon.txt",encoding='utf-8')
    lexicon=set()
    max_len=0 #lexicon中最长的词的长度
    for line in f:
        word=line.strip()
        lexicon.add(word)
        if len(word)>max_len:
            max_len=len(word)
    f.close()
    return lexicon,max_len
```

```python
lexicon,max_len=load_lexicon()
max_len
lexicon
```

```python
sentence=input("输入句子：")
sentence
```

```python
words=fmm_segment(sentence,lexicon,max_len)
```
